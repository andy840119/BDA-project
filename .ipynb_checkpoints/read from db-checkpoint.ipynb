```python
import MySQLdb
import pandas as pd
import numpy as np
conn = MySQLdb.connect(host='54.65.19.253', port=3306, user='news', passwd='newsnews', db='news')
```

```python
data = pd.read_sql('SELECT news.timestamp, title, class, gbp, xau FROM news, gbp, xau WHERE news.timestamp = gbp.timestamp AND news.timestamp = xau.timestamp;', conn)
```

```python
data.head()
```

```python
data.shape
```

```python
data = data.groupby('timestamp', as_index=False).agg({'title':lambda name: ' '.join(name), 
                                               'gbp':lambda gbp:np.mean(gbp),
                                               'xau': lambda xau:np.mean(xau)})
```

```python
data.to_csv('newstitle_gbp_xau.csv', index=False)
```
