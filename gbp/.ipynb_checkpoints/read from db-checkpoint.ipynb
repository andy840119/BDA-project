```python
import MySQLdb
import pandas as pd
import numpy as np
conn = MySQLdb.connect(host='54.65.19.253', port=3306, user='news', passwd='newsnews', db='news')
```

```python
data = pd.read_sql('SELECT news.timestamp, title, class, gbp \
                   FROM news, gbp \
                   WHERE news.timestamp = gbp.timestamp AND \
                         news.timestamp >= 1483228800;', conn)
```

```python
data.head()
```

```python
data.shape
data = data.loc[data['class'].isin(['uk-news', 'world',  'business','technology', 'money'])]
```

```python
data = data.groupby('timestamp', as_index=False).agg(
    {'title':lambda name: ' '.join(name), 
     'gbp':lambda gbp:np.mean(gbp)})
```

```python
data= data.sort_values(['timestamp'])
```

```python
data.to_csv('newstitle_gbp(uk-news,world,business,technology,money_2017-5-31).csv', index=False)
```
