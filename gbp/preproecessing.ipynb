```python
def get_date(url):
    pattern = '\d{4}/([a-z]+)/\d{2}'
    match = re.search(pattern, url)
    if match == None:
        return None
    #date = match.group()
    #datetime.strptime(match.group(), '%Y/%b/%d').timestamp()
    return match.group()
```

```python
news['date'] = news['url'].apply(get_date)
```
